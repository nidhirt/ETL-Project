### Extract Tranform Load : 


1. Extract : 
       a) Source of this data is Amazon Jobs skills dataset on Kaggle . https://www.kaggle.com/atahmasb/amazon-job-skills and Google Jobs skills dataset on Kaggle. https://www.kaggle.com/niyamatalmass/google-job-skills
          These are CSV Files in the Resources folder.
          
2. Transform :
        a) Since both the companies had their own naming conventions around Titles, Category, descriptions, location tagging, the changes were required in renaming these, normalizing them and spliting the location fields. 
        b) Split location field to city, state and country. There was "space" in the location names, which had to removed    
        c) There were various special characters which had to be removed in the description
        d) There were values in states for which countries were not given, hence had to insert such countries
        e) Renaming the countries  for uniformity in both the companies
        
3. Load :
        a) The last step was to transfer our data to DataBase. I created a database "Job skills" and respective tables to match the columns from the final Panda’s Data Frame using MYSQL and then connected to the database using SQLAlchemy and loaded the result. (SQL Queries are in queries file). By multiple queries to suit a desired criterion, the data was loaded from SQL Database
        
4. Summary: 

This database can be used to find out the maximum locations where Amazon and Google job recruitment is happening.
Which states are more likely to get Amazon and Google more hired candidates and which city is topping on the Amazon and Google chart list.
Later on it can be used to add Titles, which skillsets are Amazon and Google hiring for along with technology, basic and preferred qualifications.
